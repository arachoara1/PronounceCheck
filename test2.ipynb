```python
# [1. 텍스트 정렬]
import os
import glob
import json
import pandas as pd

# 부모 폴더 경로 설정
parent_folder = r"./test_data/child_original_text_100"  # JSON 파일이 있는 폴더 경로
output_csv_file = r"./test_data/child_original_text_100/extracted_labels.csv"  # 결과를 저장할 CSV 파일 경로

# 지원되는 파일 형식
json_extensions = ('.json')

# 폴더 내 모든 JSON 파일 불러오기
json_files = [f for f in glob.glob(os.path.join(parent_folder, "*")) if f.endswith(json_extensions)]

# 결과를 저장할 리스트
data = []

# JSON 파일 하나씩 처리
for json_file in json_files:
    print(f"처리 중: {json_file}")
    try:
        # JSON 파일 열기
        with open(json_file, "r", encoding="utf-8") as f:
            content = json.load(f)
        
        # "LabelText" 값 추출
        label_text = content.get("Transcription", {}).get("LabelText", None)
        
        # "LabelText" 값이 존재할 경우 저장
        if label_text:
            data.append(label_text)
    except Exception as e:
        print(f"오류 발생 ({json_file}): {e}")

# DataFrame으로 변환
df = pd.DataFrame(data, columns=["LabelText"])
df.index += 1  # 인덱스를 1부터 시작

# CSV 파일로 저장
df.to_csv(output_csv_file, index_label="Index", encoding="utf-8-sig")
print(f"\n모든 'LabelText'가 {output_csv_file} 파일에 저장되었습니다.")
```
